```python
#Sex type and heart disease

sns.countplot(x = 'sex', data = df, hue = 'target')
```

```python
#Males have slightly more samples without heart disease, while females have (considerably) more samples with heart disease
```

```python
#Age and heart disease

bins = [20,30,40,50,60,70,80]

df['binned_age'] = pd.cut(df['age'], bins)
sns.countplot(x = 'binned_age', data = df, hue = 'target')
```

```python
#Ages < 50 have more samples with heart disease, ages 50 -70 have more samples without heart disease, and ages > 70 again have more samples with heart disease.
```

```python
#cp (Chest pain) and heart disease

sns.countplot(x = 'cp', data = df, hue = 'target')
```

```python
#Among patients with chest pain type 0, most samples are without heart disease(over 100 out of total 138 samples with no heart disease). Most samples with chest pain type 1, 2, or 3 are with heart disease. Type 2 is the most common chest pain type among patients diagnosed with heart disease (around 70 out of total 165 diagnosed with heart disease). Almost equal numbers of those diagnosed (40) have chest pain type 0 or 1, and less than 20 have chest pain type 3. 
```

```python
#Maximum heart rate and heart disease

plt.figure(figsize = (8,4))
bins = [60,80,100,120,140,160,180,200,220]

df['binned_max_hr'] = pd.cut(df['thalach'], bins)
sns.countplot(x = 'binned_max_hr', data = df, hue = 'target')
```

```python
#Patients with maximum heart rate over 140 are more likely to be diagnosed with heart disease. This is especially true for those whose maximum heart rate is higher than 160. Further on, the higher the heart rate is, the greater is the probabilty of patient being diagnosed with heart disease.
```

```python
#Fasting blood sugar > 120 mg/dl and heart disease

sns.countplot(x = 'fbs', data = df, hue = 'target')
```

```python
#Majority of patients do not have fasting blood sugar > 120 mg/dl. Most of them (around 140) do have heart disease, while slightly less than 120 do not. Among the patients whose fbs is greater than 120 mg/dl, almost the same number (around 20) are diagnosed with heart disease.
```

```python
#Ca (number of major vessels (0-3) colored by flourosopy ) and heart disease

sns.countplot(x = 'ca', data = df, hue = 'target')
```

```python
#The smaller the number of major vessels colored by fluorosopy is, the greater probability of patient being diagnosed with heart disease.
```

```python
#Thal (defects - 1 - normal; 2 = fixed defect; 3 = reversable defect) and heart disease

sns.countplot(x = 'thal', data = df, hue = 'target')
```

```python
#Patients with thal = 2 (fixed defect, over 120 of patients in sample) are most likely to be diagnosed with heart disease, while those with thal = 3(reversable deffect) and thal = 1 (normal) are less likely to be diagnosed with heart disease. 
```

```python
#Correlation between numerical features

dfp = df[['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'target']]
sns.pairplot(dfp, hue = 'target')
```

```python
#We can see some of the features are more strongly correlated (age and resting blood pressure, age and cholesterol, resting blood presure and cholesterol, cholesterol and maximum heart rate, resting blood pressure and maxiumum heart rate, etc.)
```
