```python
#Models Implementation
```

```python
#Conversion of Categorical Variables

df = pd.get_dummies(df,columns=['sex','cp','fbs','restecg','exang','slope','thal'])
df.head()
```

```python
X = df.drop(['target', 'binned_age', 'binned_max_hr'], axis = 1)
X.head()
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = df.target, test_size = 0.2, random_state = 42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Logistic Regression Model
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
# LR Model Evaluation
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score
```

```python
#Classification Report

print(classification_report(y_test,y_pred))
```

```python
#Confusion Matrix

print(confusion_matrix(y_test, y_pred))
```

```python
#FP = 7, TN = 21, FN = 3, TP = 30
#Out of 28 instances with class 0, the classifier predicted correctly 21
#Out of 33 instancess with class 1, the classifier predicted correctly 30
```

```python
#Accuracy Score

accuracy_score(y_test,y_pred)
```

```python
#We are using a health dataset and there is a high cost associated with False Negative (people with heart desease with the wrong diagnosis). That is the main number that needs to be minimized, and recall captures exactly that. 
#The chosen metric for evaluating models will be recall
```

```python
#Recall Score
```

```python
recall_score(y_test, y_pred)
```

```python
#To visualize the performance of the classifier, the ROC Curve is used
#AUC is another measure of the classifier performance

from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import auc

probs = lr.predict_proba(X_test)
probs = probs[:,1]

fpr, tpr, thresholds = roc_curve(y_test, probs)
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr,'b', label = 'AUC = %0.2f' % roc_auc )
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('ROC Curve of Logistic Regression')
plt.show()
```

```python
#The same analysis will be performed to evaluate KNN and Random Forest classifiers
```

```python
#K-Nearest Neighbors Model (Testing for the number of neighbors that gives the best recall score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

knn = KNeighborsClassifier(n_neighbors=5)
scores = cross_val_score(knn, X, y, cv=10, scoring='recall')
print (scores.mean())

k_range = range(1, 35)
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=10, scoring='recall')
    k_scores.append(scores.mean())
print (k_scores)
```

```python
# plot the value of K for KNN versus the cross-validated recall
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Recall')
```

```python
knn = KNeighborsClassifier(n_neighbors=29)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```
